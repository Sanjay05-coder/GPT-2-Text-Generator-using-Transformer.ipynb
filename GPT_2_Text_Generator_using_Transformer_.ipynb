```python
from transformers import pipeline
```

# Download pre trained transformer model

we are using text generation model like a mino chatgpt

```python
print("Loading the brain")
generator=pipeline('text-generation',model='gpt2')
print("Brain loaded")
```

# give it a prompt

```python
input_text = "What business problem we can solve with large language models"
```

# let transformers generate output based on given input

```python
result=generator(input_text,max_length=50,num_return_sequences=1)
```

#print the output

```python
from IPython.display import display,Markdown
display(Markdown(result[0]['generated_text']))
```

#Wrap the model with gradio

```python
#define a prediction function to be used by gradio

def predict_text(prompt):
  result=generator(prompt,max_length=50,num_return_sequences=1)
  return result[0]['generated_text']
```

#create the gradio interface

```python
import gradio as gr
```

```python
interface=gr.Interface(
    fn=predict_text,
    inputs=gr.Textbox(lines=2,placeholder="Enter your prompt here"),
    outputs=gr.Textbox(lines=10),
    title="GPT 2 Text Generator",
    description="Enter a sentence to let GPT-2 complete it!"
)
```

```python
interface.launch()
```
